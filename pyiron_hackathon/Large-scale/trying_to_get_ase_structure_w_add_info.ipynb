```python
from ase.io import read
atoms = read('Ni_crack-block_1.xyz')
```

```python
from ase.lattice.cubic import FaceCenteredCubic
ase_atoms = FaceCenteredCubic(directions = [[1, 1, 2], [-1, 1, 0], [-1, -1, 1]],
                          size=(4, 4, 20),
                          symbol='Ni',
                          pbc=(1,1,0),
                         latticeconstant = 3.52)
```

```python
import numpy as np
```

```python
atom_ids = np.arange(1, len(ase_atoms) + 1)
```

```python
ase_atoms.set_array('ids', atom_ids)
```

```python
ase_atoms.get_positions()
```

```python
max_pos = np.max(ase_atoms.get_positions()[:,2])

is_bottom = ase_atoms.get_positions()[:,2] < 15
is_top = ase_atoms.get_positions()[:,2] >= max_pos - 15
is_top
```

```python
atom_flags = np.zeros(len(ase_atoms))
atom_flags[is_top] = 1
atom_flags[is_bottom] = 2
print(atom_flags)
```

```python
import numpy as np
from contextlib import contextmanager

@contextmanager
def show_complete_array():
    oldoptions = np.get_printoptions()
    np.set_printoptions(threshold=np.inf)
    try:
        yield
    finally:
        np.set_printoptions(**oldoptions)
```

```python
ase_atoms.set_array('flags', atom_flags)
```
