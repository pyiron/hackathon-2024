```python
from pyiron_workflow import Workflow as wf
from typing import Optional
```

Let's start with writing some trivial nodes for getting started

```python
@wf.wrap.as_function_node()
def addOne(
    x: Optional[float | int] = None
):
    y = x + 1
    return y
```

```python
@wf.wrap.as_function_node()
def addTwo(
    x: Optional[float | int] = None
):
    y = x + 2
    return y
```

```python
wf_add = wf("test")
wf_add.firstAddition = addOne(x = 4.)
wf_add.secondAddition = addTwo(x = wf_add.firstAddition)
```

```python
wf_add.draw(size=(10,10))
```

```python
wf_add.secondAddition.pull()
```
