```python
from pyiron_workflow import Workflow
from typing import Optional
import numpy as np
```

```python
%config IPCompleter.evaluation='unsafe'
```

## Dataclasss nodes

```python
@Workflow.wrap.as_dataclass_node
class InputDFT:
    import numpy
    kPoints: float| list | numpy.ndarray
    enCut: float | list | numpy.ndarray
    strain: float | list | numpy.ndarray
    pr_name: str

@Workflow.wrap.as_dataclass_node
class MaterialInput:
    element: str = 'Li'
    bravais_basis: Optional[str] = None
    a: float = 3.0
    b: Optional[str] = None
    c: Optional[str] = None
    cubic: bool = False
```

## function nodes

```python
@Workflow.wrap.as_function_node("bulk_structure")
def structure_creation(input: MaterialInput.dataclass):
    ''' Creates a bulk structure '''
    from pyiron_atomistics import Project
    pr = Project('.')    
    bulk_structure = pr.create_structure(element=input.element, bravais_basis=input.bravais_basis, lattice_constant=input.a)
    return bulk_structure
    

@Workflow.wrap.as_function_node("strained_structure")
def apply_strain(structure, strain, copy_structure = True):
    '''apply strain to the structure '''
    if not copy_structure:
        strained_structure = structure.set_cell(cell=structure.cell * strain ** (1/3), scale_atoms=True)
    strained_structure = structure.copy()
    strained_structure.set_cell(cell=structure.cell * strain ** (1/3), scale_atoms=True)
    return strained_structure

@Workflow.wrap.as_function_node("sphinx_strain_job")
def sphinx_calculation_for_encut_strain(pr_name, structure, encut, strain):
    """ Function to set sphinx jobs with varying strain on the lattice. 
    Takes inputs project, basis, strain and the cutoff energy (encut) """
    from pyiron_atomistics import Project
    pr = Project(pr_name)
    encut_str = str(encut).replace('.','_')
    strain_str = str(strain).replace('.', '_')
    job_sphinx_strain = pr.create_job(job_type=pr.job_type.Sphinx, job_name='sphinx_encut' + encut_str + 'strain' + strain_str)
    job_sphinx_strain.structure = structure
    job_sphinx_strain.set_occupancy_smearing('Fermi', width=0.1)
    job_sphinx_strain.set_encut(encut)    
    queue = 'cmti'
    job_sphinx_strain.server.queue = queue
    job_sphinx_strain.server.cores = 1
    job_sphinx_strain.run(delete_existing_job=True)
    return job_sphinx_strain


@Workflow.wrap.as_function_node("sphinx_single_atom_job")
def single_atom_encut(pr_name,encut,structure):
    '''function to do single atom calculations for encut variation'''
    from pyiron_atomistics import Project
    pr = Project(pr_name)
    encut_str = str(encut).replace('.','_')
    job_sa = pr.create_job(job_type=pr.job_type.Sphinx, job_name='sa_' + encut_str)
    job_sa.set_occupancy_smearing('Fermi', width=0.1)
    job_sa.set_encut(encut)
    job_sa.structure = structure
    queue = 'cmti'
    job_sa.server.queue = queue
    job_sa.server.cores = 2
    job_sa.run()
    return job_sa

@Workflow.wrap.as_function_node("volume_dict","energy_dict")
def energy_volume_extractor(pr_name, job_df):
    '''function to do extract energy and volume'''
    from pyiron_atomistics import Project
    import re
    from collections import defaultdict
    pr = Project(pr_name)
    volume_dict = defaultdict(dict)
    energy_dict = defaultdict(dict)
    for job in job_df['sphinx_strain_job']:
        pr.wait_for_job(job,max_iterations=100_000)
        pattern = r"encut(\d+)strain(\d+)_(\d+)"
        match = re.search(pattern, job.job_name)
        encut = int(match.group(1))  # Extracts the encut value
        strain = float(f"{match.group(2)}.{match.group(3)}") 
        volume_dict[encut][strain] = job['output/generic/volume'][-1]
        energy_dict[encut][strain] = job['output/generic/energy_pot'][-1]      
    return volume_dict, energy_dict

@Workflow.wrap.as_function_node("volume_fit_dict","energy_fit_dict", "min_energy_dict","min_volume_dict")
def murnaghan_job(pr_name, volume_dict, energy_dict):
    import numpy as np
    from pyiron_atomistics import Project
    
    pr = Project(pr_name)
    volume_fit_dict = {}
    energy_fit_dict = {}
    min_energy_dict = {}
    min_volume_dict = {}
    for key in volume_dict.keys():
        murn = pr.create_job(
            job_type=pr.job_type.Murnaghan, job_name='murn' + str(pr.base_name)
        )
        [e0, b0, bP, v0], [e0_error, b0_error, bP_error, v0_error] = murn._fit_leastsq(
            volume_lst=list(volume_dict[key].values()), energy_lst=list(energy_dict[key].values()), fittype='birchmurnaghan'
        )
        # Fit curve
        volume_fit_dict[key] = np.linspace(np.min(list(volume_dict[key].values())), np.max(list(volume_dict[key].values())), 1000)
        fit_energy = murn.fit_module.birchmurnaghan_energy(list(volume_fit_dict[key]), e0, b0 / 160.21766208, bP, v0)
        energy_fit_dict[key] = fit_energy
        min_energy_dict[key] = np.min(fit_energy)
        min_volume_dict[key] = volume_fit_dict[key][np.argmin(fit_energy)]
        
    return volume_fit_dict, energy_fit_dict, min_energy_dict, min_volume_dict

@Workflow.wrap.as_function_node("figure")
def subplots(x: dict, y: dict, n_rows: int, n_cols: int, xlabel: str, ylabel:str ):
    import matplotlib.pyplot as plt
    i=1
    fig = plt.figure(figsize=(14,10))
    assert n_rows*n_cols>=len(x.keys()),"The number of subplots is less than the number of data keys"
    for key in x.keys():
        plt.subplot(n_rows, n_cols, i)
        plt.plot(x[key], y[key], 'x-', label=key)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.legend()
        i+=1
    plt.tight_layout()
    return fig
```

## Macro nodes

```python
@Workflow.wrap.as_macro_node("sphinx_strain_job")
def MurnaghanSphinx(wf, element, bravais_basis, a , strain, encut,  pr_name):
    wf.material = MaterialInput(element=element,bravais_basis=bravais_basis,a=a)
    wf.structure_creation = structure_creation(input=wf.material)
    wf.apply_strain = apply_strain(structure=wf.structure_creation, strain=strain)
    wf.sphinx_strain_job = sphinx_calculation_for_strain(pr_name=pr_name, structure = wf.apply_strain, encut=encut, strain=strain) 
    return wf.sphinx_strain_job

@Workflow.wrap.as_macro_node("volume_fit_dict","energy_fit_dict",'figure')
def EnergyVolumeMurnaghanSphinx(wf, material: MaterialInput.dataclass, k_encut_inp: InputDFT.dataclass):
    wf.job_df = Workflow.create.for_node(
        body_node_class = MurnaghanSphinx,
        iter_on = ("strain","encut"),
        strain = k_encut_inp.strain,
        encut = k_encut_inp.enCut,
        element = material.element,
        bravais_basis = material.bravais_basis,
        a = material.a,
        pr_name = k_encut_inp.pr_name
    )
    wf.energy_volume_extraction = energy_volume_extractor(
        pr_name = k_encut_inp.pr_name, 
        job_df = wf.job_df.outputs.df
    )
    wf.murnaghan_fit = murnaghan_job(
        pr_name = k_encut_inp.pr_name, 
        volume_dict = wf.energy_volume_extraction.outputs.volume_dict, 
        energy_dict = wf.energy_volume_extraction.outputs.energy_dict
    )
    wf.plots = subplots(x = wf.murnaghan_fit.outputs.volume_fit_dict, y = wf.murnaghan_fit.outputs.energy_fit_dict, n_rows=5, n_cols=3, xlabel='Volume ($\AA ^ 3$)', ylabel='Energy (eV)' )
    return wf.murnaghan_fit.outputs.volume_fit_dict, wf.murnaghan_fit.outputs.energy_fit_dict, wf.plots.outputs.figure
```
